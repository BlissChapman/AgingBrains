```python
import os
import pandas
from skimage import io
from tqdm import tqdm
```

```python
files = os.listdir('raw/')
```

```python
attributes = pandas.read_csv('list_attr_celeba.csv')
attributes = attributes.set_index('image_id')
attributes.head()
```

# Age 

```python
for filename in tqdm(files):
    is_young = attributes.loc[filename]['Young']
    class_dir = 'young/{}/'.format(is_young)
    if not os.path.exists(class_dir):
        os.mkdir(class_dir)
    img = io.imread('raw/' + filename)
    io.imsave(class_dir + filename, img)
```

# Gender

```python
for filename in tqdm(files):
    is_male = attributes.loc[filename]['Male']
    class_dir = 'male/{}/'.format(is_male)
    if not os.path.exists(class_dir):
        os.mkdir(class_dir)
    img = io.imread('raw/' + filename)
    io.imsave(class_dir + filename, img)
```
